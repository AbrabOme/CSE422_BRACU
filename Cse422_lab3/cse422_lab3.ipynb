```python
import random
import math


def maximini(graph,state, alpha, beta, booli):
    if type(state) == type(0):
        return state
    if booli:
        max_val =- math.inf
        for itr in graph[state]:
            eval = maximini(graph, itr, alpha, beta, False)
            max_val = max(max_val,eval)
            alpha = max(alpha,eval)
            if beta <= alpha:
                break
        return max_val
    else:
        min_val = math.inf
        for itr in graph[state]:
            eval = maximini(graph,itr, alpha, beta, True)
            min_val = min(min_val, eval)
            beta = min(beta, eval)
            if beta<=alpha:
                break
        return min_val

stu_id = str(input("Enter Your Student Id: "))

min_point = int(stu_id[4])

points_to_win = int(stu_id[7]+stu_id[6])
max_point = int(points_to_win*1.5)

shuffle = int(stu_id[3])
if shuffle == 0:
    shuffle = 8

graph = {"A0":["A1","B1"],"A1":["A2","B2"],"B1":["C2","D2"],"A2":[],"B2":[], "C2":[],"D2":[]}
optimus_count = 0
final_list = []

lst = []
for i in range(8):
    v = random.randint(min_point, max_point)
    lst.append(v)

graph["A2"].append(lst[0])
graph["A2"].append(lst[1])
graph["B2"].append(lst[2])
graph["B2"].append(lst[3])
graph["C2"].append(lst[4])
graph["C2"].append(lst[5])
graph["D2"].append(lst[6])
graph["D2"].append(lst[7])

ans = maximini(graph,"A0",-math.inf, math.inf, True)
final_list.append(ans)
print("Generated 8 random points between the minimum and maximum point limits:",lst)
print("Total points to win:", points_to_win)
print("Achieved point by applying alpha-beta pruning =",ans)
if ans>= points_to_win:
    print("The Winner is Optimus Prime")
    optimus_count += 1
else:
    print("The Winner is Megatron")


for i in range(shuffle-1):
    lst.append(lst.pop(0))
    graph["A2"].append(lst[0])
    graph["A2"].append(lst[1])
    graph["B2"].append(lst[2])
    graph["B2"].append(lst[3])
    graph["C2"].append(lst[4])
    graph["C2"].append(lst[5])
    graph["D2"].append(lst[6])
    graph["D2"].append(lst[7])

    ans = maximini(graph,"A0",-math.inf, math.inf, True)
    final_list.append(ans)
    if ans >= points_to_win:
        optimus_count += 1
print("========================================")
print("After the shuffle:")
print("List of all points values from each shuffle:",final_list)
print("The maximum value of all shuffles:",max(final_list))
print(f"Won {optimus_count} times out of {shuffle} number of shuffles")
```
