```python
import random
input = open("input.txt")
line = input.readline().split(" ")
player_no = int(line[0])
total_run = int(line[1])

arr = []
final = []
player_list = []

for i in range(player_no):
    arr1 = input.readline().split(" ")
    name = arr1[0]
    p_run = int(arr1[1])
    player_list.append((name,p_run))
#print(player_list)
def fitness(arr2):
    check_sum = 0
    for i in range(player_no):
        if arr2[i] == 1:
            check_sum += player_list[i][1]

    return check_sum

limit = 0
while limit < 50:
    chrom = []
    for i in range(player_no):
        chrom.append(random.randint(0,1))
        if chrom not in arr:
            arr.append(chrom)
            limit += 1

def crossover(parent1,parent2):
    p1_head = parent1[0:player_no//2]
    p1_tail = parent1[player_no//2:]
    p2_head = parent1[0:player_no//2]
    p2_tail = parent1[player_no//2:]


    child1 = p1_head + p2_tail
    child2 = p2_head + p1_tail

    return [child1,child2]

def mutation(child):
    rando = random.randint(0,player_no-1)
    if  child[rando] == 1:
        child[rando] = 0
    elif  child[rando] == 0:
        child[rando] = 1
    return child


def goal_check(parent1,parent2):

    answer = []
    fit_check1 = fitness(parent1)
    fit_check2 = fitness(parent2)

    if fit_check1 == total_run:
        answer.append(parent1)

    elif fit_check2 == total_run:
        answer.append(parent2)
    else:
        return 0
    return answer

for i in range(10000):

    random.shuffle(arr)
    if len(arr)>= 2:
        first = arr.pop()
        second = arr.pop()

    ans = goal_check(first,second)
    if ans == 0:
        counter = random.random()
        child = crossover(first,second)
        baccha_1 = child[0]
        baccha_2 = child[1]

        if counter >= 0.6:
            x1 = mutation(baccha_1)
            x2 = mutation(baccha_2)

            ans = goal_check(x1,x2)

            if ans == 0:
                if x1 not in arr:
                    arr.append(x1)
            else:
                break

            if ans == 0:
                if x2 not in arr:
                    arr.append(x2)
            else:
                break

        else:
            ans = goal_check(baccha_1,baccha_2)
            if ans == 0:
                if baccha_1 not in arr:
                    arr.append(baccha_1)
                if baccha_2 not in arr:
                    arr.append(baccha_2)
            else:
                break
    else:
        break
lst=[]
for i in range(len(player_list)):
    lst.append(player_list[i][0])
print(lst)
for i in ans[0]:
    print(i,end="")






```
